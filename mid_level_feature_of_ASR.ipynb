```python
def get_rename_id_dict():
    ''' rename_id_dict
  'Gong_8_4': 'Gong_8_4_ES','''
    rename_dict=np.load('all_cilp_rename.npy', allow_pickle=True).item()
    rename_id_dict = {}
    for key in rename_dict.keys():
        key_id = key[:-4]
        value_id = rename_dict[key][:-4]
        rename_id_dict[key_id] = value_id
    return rename_id_dict
```

```python
def summary_df(asr_df):
    for i in range(len(asr_df)):
        one_turn = asr_df.loc[i,'asr']
        [turn_id,asr_txt]=one_turn.split(' ',1)
        [a, teacher, lecture, clip, turn, start, end] = turn_id.split('_')
#         print(turn_id)
        start_time = float(start)/100
        end_time = float(end)/100
        # 单位秒 736：7.36秒
        time_gap = end_time-start_time
    #     print(one_turn)
    #     print('=====,',asr_txt.split())
    #     print(len(asr_txt))
        if len(asr_txt) >0:
            if len(asr_txt.split()) >0:
                if (asr_txt.split())[0] != '<UNK>':
                    asr_txt_no = asr_txt.replace(' ','')
                    speed = len(asr_txt_no)/time_gap
        #         print(speed)
                else:
                    speed = None
        else:
            speed = None

        clip_id = teacher +'_'+lecture+'_'+clip
        # 'Gong_10_0'
        asr_ls = asr_txt.split()
        
        if clip_id in big_type_one_dict.keys():
        
            clip_big_type = big_type_one_dict[clip_id]
        
        word_count = len(asr_txt.split())
        
        asr_df.loc[i,'word_amount'] = word_count
        asr_df.loc[i,'clip_id'] = clip_id
        asr_df.loc[i,'turn'] = turn
        asr_df.loc[i,'start_time'] = start_time
        asr_df.loc[i,'end_time'] = end_time
        asr_df.loc[i,'speed'] = speed
        asr_df.loc[i,'asr_txt'] = asr_txt
#         asr_df.loc[i,'clip_type'] = clip_type
        asr_df.loc[i,'clip_big_type'] = clip_big_type
```

```python
rename_dict=get_rename_id_dict()
big_type_one_dict = np.load('big_type_one_dict.npy', allow_pickle=True).item()
```

```python
def get_mean_dict(ori_dict):
    mean_dict = {}
    for key in list(ori_dict.keys()):
        mat = ori_dict[key]
        vec = np.mean(mat, axis=0)
        mean_dict[key] = vec
    return mean_dict

def get_sum_dict(ori_dict):
    sum_dict = {} 
    for key in list(ori_dict.keys()):
        sum_dict[key] = sum(ori_dict[key])
    return sum_dict

def get_max5_dict(ori_dict):
    max5_dict = {} 
    for key in list(ori_dict.keys()):
        ls = sorted(ori_dict[key],reverse = True) 
        if len(ls) <5:
            mean = np.mean(ls)
            while len(ls) <5:
                ls.append(mean)
            
            new_ls = sorted(ls,reverse=True)
            max5_dict[key] = new_ls
#             print(key,len(ls))
        else:
            
            max5_dict[key] = ls[:5]
    return max5_dict
```

```python
def get_speed_array(asr_df,clip_big_type):
    '''
    get speech speed array of a clip
    '''
    mini_df = asr_df[asr_df['clip_big_type'].isin([clip_big_type])]
    speed_ls = np.array((mini_df['speed']))
    new_speed_ls = [x for x in speed_ls if np.isnan(x) == False]
    return new_speed_ls

def get_asr_speed_dict(asr_df):
    asr_speed_dict={}

    for clip_big_type in set(asr_df['clip_big_type']):
        new_speed_ls = get_speed_array(asr_df,clip_big_type)
        if len(new_speed_ls) <1:
            print(clip_big_type)
            print(new_speed_ls)
        asr_speed_dict[clip_big_type] =  new_speed_ls
    return asr_speed_dict
```

```python
def get_count_array(asr_df,clip_big_type):
    mini_df = asr_df[asr_df['clip_big_type'].isin([clip_big_type])]
    speed_ls = np.array((mini_df['word_amount']))
    new_ls = [x for x in speed_ls if np.isnan(x) == False]
    return new_ls
def get_asr_count_dict(asr_df):
    asr_count_dict={}
    for clip_big_type in set(asr_df['clip_big_type']):
        new_ls = get_count_array(asr_df,clip_big_type)
        if len(new_ls) <1:
            print(clip_big_type)
            print(new_ls)
        asr_count_dict[clip_big_type] =  new_ls
    return asr_count_dict
```

```python
def get_token_count(asr_df):

    token_set_dict = {}
    token_mean_dict = {}
    for clip_big_type in set(asr_df['clip_big_type']):
        mini_df = asr_df[asr_df['clip_big_type'].isin([clip_big_type])]


        clip_ls = list(mini_df['asr_txt'])
        clip_str = ''.join(clip_ls)
        token_ls = clip_str.split()
#         print(token_ls)
        token_set = set(token_ls)
        token_set_dict

        token_count = []
        for clip in clip_ls:
            clip_token = len(set(clip_str.split(clip)))
            token_count.append(clip_token)
        token_mean = np.mean(token_count)

        token_mean_dict[clip_big_type] = token_mean
        token_set_dict[clip_big_type] = len(token_set)
    return token_set_dict,token_mean_dict
def get_TTR(token_set_dict,asr_count_sum_dict):
    '''
    TTR = NDW/TNM
    NDW: token_set_dict
    TNM: asr_count_sum_dict
    '''
    asr_TTR_dict = {}
    for key in token_set_dict.keys():
        NDW = token_set_dict[key] 
        TNM = asr_count_sum_dict[key]
        TTR = NDW/TNM
        asr_TTR_dict[key] = TTR
    return asr_TTR_dict
```

```python
asr_speed_dict = get_asr_speed_dict(asr_df)
asr_speed_mean_dict = get_mean_dict(asr_speed_dict)

asr_count_dict = get_asr_count_dict(asr_df)
asr_count_mean_dict = get_mean_dict(asr_count_dict)
asr_count_sum_dict = get_sum_dict(asr_count_dict)
asr_count_max5_dict = get_max5_dict(asr_count_dict )

token_set_dict,token_mean_dict = get_token_count(asr_df)

asr_ttr_dict = get_TTR(token_set_dict,asr_count_sum_dict)
```
